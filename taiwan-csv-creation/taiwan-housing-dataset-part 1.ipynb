```python
import os
import pandas as pd

newtaipeidf = pd.read_excel(r'/home/arberaga/Downloads/New_Taipei_City/merged.xlsx')
taipeidf = pd.read_excel(r'/home/arberaga/Downloads/Taipei/merged.xlsx')
taoyuandf = pd.read_excel(r'/home/arberaga/Downloads/Taoyuan/merged.xlsx')
taichungdf = pd.read_excel(r'/home/arberaga/Downloads/Taichung/merged.xlsx')
```

```python
newtaipeidf_split = newtaipeidf.drop(axis=1,labels=["Unnamed: 18_level_1","Unnamed: 19_level_1","Unnamed: 0","車位總價(萬元)"])
newtaipeidf_split["City"] = "New Taipei"
taipeidf_split = taipeidf.drop(axis=1,labels=["Unnamed: 18_level_1","Unnamed: 19_level_1","Unnamed: 0","車位總價(萬元)"])
taipeidf_split["City"] = "Taipei"
taoyuandf_split = taoyuandf.drop(axis=1,labels=["Unnamed: 18_level_1","Unnamed: 19_level_1","Unnamed: 0","車位總價(萬元)"])
taoyuandf_split["City"] = "Taoyuan"
taichungdf_split = taichungdf.drop(axis=1,labels=["Unnamed: 18_level_1","Unnamed: 19_level_1","Unnamed: 0","車位總價(萬元)"])
taichungdf_split["City"] = "Taichung"

#removed the column parking space as it contains only 2 values in 350K rows
```

```python
all_df_split = pd.concat([newtaipeidf_split,taipeidf_split,taoyuandf_split,taichungdf_split])
all_df_split.sort_values("交易日期",inplace=True)	
all_df_split.rename(inplace=True,columns={"地段位置或門牌":"Lot location or house number","社區簡稱":"Community abbreviation","交易日期":"transaction date","總價(萬元)":"Total price (10,000 yuan)","單價(萬元/坪)":"Unit price (10,000 yuan/square meter)","總面積(坪)":"Total area(ping)","主建物佔比":"Proportion of main building","型態":"type","屋齡":"House age","樓別/樓高":"Floor type/floor height","交易標的":"Transaction object","交易筆棟數":"Number of transactions","建物現況格局":"The current layout of the building","管理組織":"management organization","電梯":"elevator","主要用途":"The main purpose","備註":"Remark"})
# DONT drop rows which have the column Proportion of main building & The current layout of the building as NaN,
# all_df_split.dropna(subset=['Proportion of main building','The current layout of the building'],inplace=True)
all_df_split.reset_index(inplace=True)
all_df_split.drop(labels=["index","Transaction object"],axis=1,inplace=True)
all_df_split[all_df_split["Total price (10,000 yuan)"].isna()]
all_df_split
```

```python
trainmask = (all_df_split['transaction date'] < "110/01/01")
#validatemask = (all_df_split['transaction date'] >= "110/01/01") & (all_df_split['transaction date'] <= "110/03/31")
testmask = (all_df_split['transaction date'] >= "110/01/01")
train_df = all_df_split.loc[trainmask]
test_df = all_df_split.loc[testmask]
#validate_df = all_df_split.loc[validatemask]
#duplicates will not be removed
```

```python
all_df_split["Proportion of main building"] = all_df_split["Proportion of main building"].fillna(train_df["Proportion of main building"])
```

```python
all_df_split["House age"] = all_df_split["House age"].fillna(train_df["House age"].median())
```

```python
all_df_split["The main purpose"] = all_df_split["The main purpose"].fillna("住家用")
```

```python
all_df_split["The current layout of the building"] = all_df_split["The current layout of the building"].fillna("3房2廳2衛")
```

```python
all_df_split["Number of transactions"] = all_df_split["Number of transactions"].fillna("土:1 建:1車:0")
```

```python
all_df_split["Floor type/floor height"] = all_df_split["Floor type/floor height"].fillna("全/四層")
```

```python
#split "number of transactions" column into 3 different columns based on semicolon of column
# x = "土:1 建:1車:0"
# arr = re.findall("(\d)",x)
import re

def handle(x,chars):

    val = re.findall(chars+":(\d*)",x)
    if val == []:
        val = [0]
    return val[0]

def handle2(x,chars):

    val = re.findall("(\d*)"+chars,x)
    if val == []:
        val = [0]
    return val[0]


all_df_split["Land"] = all_df_split["Number of transactions"].apply(lambda x : handle(x, "土"))
all_df_split["Construction"] = all_df_split["Number of transactions"].apply(lambda x : handle(x, "建"))
all_df_split["Car"] = all_df_split["Number of transactions"].apply(lambda x : handle(x, "車"))

all_df_split["Bedrooms"] = all_df_split["The current layout of the building"].apply(lambda x : handle2(x, "房"))
all_df_split["Living rooms"] = all_df_split["The current layout of the building"].apply(lambda x : handle2(x, "廳"))
all_df_split["Bathrooms"] = all_df_split["The current layout of the building"].apply(lambda x : handle2(x, "衛"))
all_df_split["elevator"].fillna("無")
all_df_split["elevator"].replace({"無":0,"有":1},inplace=True)
all_df_split["management organization"].fillna("無")
all_df_split["management organization"].replace({"無":0,"有":1},inplace=True)
all_df_split["management organization"].describe()
```

```python
#Creating external nodes and edges
ordinal_enc = OrdinalEncoder()

all_df_split['Encoded_type'] = ordinal_enc.fit_transform(all_df_split[['type']]).astype(int)
Encoded_type = all_df_split[["type", "Encoded_type"]].drop_duplicates().sort_values(by='Encoded_type')

all_df_split['Encoded_floor'] = ordinal_enc.fit_transform(all_df_split[['Floor type/floor height']]).astype(int)
Encoded_floor = all_df_split[["Floor type/floor height", "Encoded_floor"]].drop_duplicates().sort_values(by='Encoded_floor')

all_df_split['Encoded_purpose'] = ordinal_enc.fit_transform(all_df_split[['The main purpose']]).astype(int)
Encoded_purpose = all_df_split[["The main purpose", "Encoded_purpose"]].drop_duplicates().sort_values(by='Encoded_purpose').reset_index().drop(columns="index")


Encoded_floor = Encoded_floor.dropna(subset=["Floor type/floor height"]).reset_index().drop(columns="index")
Encoded_purpose = Encoded_purpose.dropna(subset=["The main purpose"]).reset_index().drop(columns="index")
Encoded_type.to_csv('nodes_type_data.csv')
Encoded_floor.to_csv('nodes_floor_data.csv')
Encoded_purpose.to_csv('nodes_purpose_data.csv')


type_data = all_df_split["Encoded_type"].copy()
floor_data = all_df_split["Encoded_floor"].copy()
purpose_data = all_df_split["Encoded_purpose"].copy()


type_data = type_data.reset_index().rename(columns={"index":"node_id"})
floor_data = floor_data.reset_index().rename(columns={"index":"node_id"})
purpose_data = purpose_data.reset_index().rename(columns={"index":"node_id"})
print(len(purpose_data))
type_data = type_data[type_data["Encoded_type"]>=0]
floor_data = floor_data[floor_data["Encoded_floor"]>=0]
purpose_data = purpose_data[purpose_data["Encoded_purpose"]>=0]

print(len(purpose_data))

type_data.to_csv('edges_type_data.csv')
floor_data.to_csv('edges_floor_data.csv')
purpose_data.to_csv('edges_purpose_data.csv')

all_df_split['Encoded_city'] = ordinal_enc.fit_transform(all_df_split[['City']]).astype(int)
Encoded_city = all_df_split[["City", "Encoded_city"]].drop_duplicates().sort_values(by='Encoded_city').reset_index().drop(columns="index")
Encoded_city.to_csv('nodes_city_data.csv')
city_data = all_df_split["Encoded_city"].copy()
city_data = city_data.reset_index().rename(columns={"index":"node_id"})
city_data = city_data[city_data["Encoded_city"]>=0]

city_data.to_csv('edges_city_data.csv')

all_df_split_removed =all_df_split.drop(columns=["Community abbreviation","type","Floor type/floor height","The main purpose", "Number of transactions", "The current layout of the building", "Remark"])
```
